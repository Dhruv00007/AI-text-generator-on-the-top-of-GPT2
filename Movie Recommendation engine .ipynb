```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv(r'C:\Users\Welcome\Desktop\Github Data\movies.csv')
```

```python
df['Movie_id'] = range(0,1000)
```

```python
df.head(5)
```

```python
df.shape
```

```python
columns = ['Actors','Director','Genre','Title']
```

```python
df[columns].head(3)
```

```python
#checking for missing values
df[columns].isnull().values.any()
```

```python
#creating function for important features into one string
def get_important_features(data):
    important_features = []
    for i in range(0,data.shape[0]):
        important_features.append(data['Actors'][i]+' '+data['Director'][i]+' '+data['Genre'][i]+' '+data['Title'][i])
    return important_features
```

```python
#holding data in a single column 
df['important_features'] = get_important_features(df)

df.head(3)
```

```python
#converting the data in matrix of data count 
cm = CountVectorizer().fit_transform(df['important_features'])
```

```python
#checking for similarity 
cs = cosine_similarity(cm)
print(cs)
```

```python
cs.shape
```

```python
#getting the title of movie that users like 
title = 'The Amazing Spider-Man'

#finding movie id with title the amazing spider man
movie_id = df[df.Title==title]['Movie_id'].values[0]
```

```python
#enumeration for similarity score 
scores = list(enumerate(cs[movie_id]))
print(scores)
```

```python
#sorting the list here lambda is used for checking similarity 
sorted_scores = sorted(scores,key = lambda x:x[1] , reverse=True)
sorted_scores = sorted_scores[1:]
```

```python
print(sorted_scores)
```

```python
#creating loop for first 7 similar movies 
j=0
print('The 7 most recommended movies to',title,'are:\n')
for item in sorted_scores:
    movie_title = df[df.Movie_id == item[0]]['Title'].values[0]
    print(j+1,movie_title)
    j = j+1
    if j>6:
        break
```
